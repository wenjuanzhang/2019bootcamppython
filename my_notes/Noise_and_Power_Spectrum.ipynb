# Noise and Power Spectrum

We are going to generate and graph two kinds of noise, white and brown, then examine (and explain) their *power spectra*.  

[Listen to white noise](https://en.wikipedia.org/wiki/File:White-noise-sound-20sec-mono-44100Hz.ogg#file)

[Listen to brownian noise](https://en.wikipedia.org/wiki/File:Brownnoise.ogg)

[About white noise (Wikipedia)](https://en.wikipedia.org/wiki/White_noise)

[About brownian noise (Wkipedia)](https://en.wikipedia.org/wiki/Brownian_noise)

```python
# We will generate, then graph some white noise:

import random as r

wn = []
for i in range(0,1000): 
    wn.append(r.random())

import matplotlib.pyplot as plt
plt.plot(wn)
```

```python
# Next, we will compute the power spectrum of the noise
import numpy as np

wnff = np.fft.fft(wn)
wna = np.absolute(wnff)
wnp = wna*wna
plt.plot(wnp[1:])
```

```python
# A few things to notice.  First, the spectrum looks
# somewhat like a constant function with value
# about 50 + moderate noise + a few spikes.  Second,
# Second, the graph appears symmetric with respect
# to reflecton about the x-midline, x = 500.  We
# will explain this later. It is a consequence of
# the fact that the data consists of real (not complex)
# numbers.
```

```python
# Now we will generate some Brownian noise.
import numpy as np
import random as r
import matplotlib.pyplot as plt
%matplotlib inline
bn = [0]
for i in range(0,999):
    bn.append(bn[i] + 0.1*(r.random() - 0.5))
    
plt.plot(bn)
```

## Discrete Fourier Transform

```python
bnff = np.fft.fft(bn) 
bna = np.absolute(bnff) 
bnp = bna*bna 
plt.plot(bnp[40:-40])
plt.figure()
plt.plot(np.log(bnp[0:-500]))
```

```python
xs = range(1,500)
ps = bnp[1:500]
len(xs), len(ps)
```

```python
lxs = np.log(xs)
lys = np.log(ps)
plt.plot(lxs, lys)
```

```python
np.polyfit(lxs, lys, 1)
```
